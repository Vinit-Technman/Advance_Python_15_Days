Create a function that generates a time series 
decomposition using seasonal decomposition of time 
series (STL) in Python

```python
# https://www.geeksforgeeks.org/seasonal-decomposition-of-time-series-by-loess-stl/
# Seasonal-Trend decomposition using Loess (STL) is a specific 
# decomposition method that employs the Loess technique to separate a time series into its trend, seasonal, and residual components.
from statsmodels.tsa.seasonal import STL
import matplotlib.pyplot as plt
df=pd.read_csv("Tesla_Stock.csv",index_col="Date",parse_dates=True)
df
ax=plt.figure(figsize=(12,4))
plt.plot(df.index,df['Close'],color='green')
plt.show()
```

```python
stl=STL(df['Close'],seasonal=11,robust=True,period=7)
result=stl.fit()

# result
fig, (ax1, ax2, ax3,ax4) = plt.subplots(4, 1, figsize=(10, 6))
ax4.plot(df.index,df['Close'],color='green')
ax4.set_title("Original Component")
ax1.plot(df.index, result.trend, label='Trend', color='red')
ax1.set_title('Trend Component')
ax2.plot(df.index, result.seasonal, label='Seasonal', color='blue')
ax2.set_title('Seasonal Component')
ax3.plot(df.index, result.resid, label='Residual')
ax3.set_title('Residual Component')
plt.tight_layout()

plt.show()
```

 Implement a program that uses Pandas to detect and 
remove outliers in a time series and visualize the results 
using Matplotlib.

```python
import matplotlib.pyplot as plt
from scipy.stats import zscore
# Q1 = df['Height'].quantile(0.25)
# Q3 = df['Height'].quantile(0.75)
# IQR = Q3 - Q1

# # identify outliers
# threshold = 1.5
# outliers = df[(df['Height'] < Q1 - threshold * IQR) | (df['Height'] > Q3 + threshold * IQR)]
# we;re finding outliers usign zscores
df=pd.read_csv("Tesla_Stock.csv",index_col="Date",parse_dates=True)
df["zscore"]=zscore(df['Close'])
threshold_val=1
df['outlier']=df['zscore'].abs()>threshold_val

# plt.boxplot(df['zscore'])
# plt.show()
plt.plot(df.index,df['Close'],label='Original')
plt.scatter(df[df['outlier']].index,df[df['outlier']]['Close'],color='red',label='Outliers')
plt.legend()
plt.title("ORiginal Time Series With outliers")
plt.show()

# remove those outliers
df_cleaned = df[~df['outlier']]
plt.plot(df_cleaned.index,df_cleaned['Close'],label='Cleaned')
plt.legend()
plt.title("Modi Time Series Without outliers")
plt.show()
df
```
