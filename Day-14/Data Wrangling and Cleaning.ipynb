Drop missing values from a DataFrame using Pandas

```python
import pandas as pd
import numpy as np

d1 = {
'Name': ['Shark', 'Whale', 'Jellyfish', 'Starfish'],
'ID': [1, 2, 3, 4],
'Population': [100, 200, np.nan, pd.NaT],
'Regions': [1, None, pd.NaT, pd.NaT]
}

df1 = pd.DataFrame(d1)
print(df1)

df2=df1.dropna()
print(df2)
```

 Remove duplicate rows from a DataFrame using Pandas.

```python
d1 = {
'Name': ['Shark', 'Whale', 'Jellyfish', 'Shark'],
'ID': [1, 2, 3, 1],
'Population': [100, 200, np.nan, 100],
'Regions': [1, None, pd.NaT, 1]
}

df1 = pd.DataFrame(d1)
df1

print(df1)

print()
df2=df1.drop_duplicates(keep='first')
print(df2)
```

Merge two DataFrames using an outer join and handle 
missing values appropriately

```python
df1 = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Product': ['Apples', 'Bananas', 'Cherries']
})
df2 = pd.DataFrame({
    'CustomerID': [2, 3, 4],
    'Quantity': [3, 5, 2]
})

display(df1)
display(df2)

df=pd.merge(df1.set_index("CustomerID"),df2.set_index("CustomerID"),on='CustomerID',how='outer')

display(df)

df["Quantity"]=df["Quantity"].fillna(df["Quantity"].mean())

display(df)
```

Aggregate data in a DataFrame to calculate the mean, 
median, and standard deviation for each group

```python
df=pd.read_csv("Employees_Cleaned.csv")
df.drop("Unnamed: 0",axis=1,inplace=True)

display(df)

salary_group=df.groupby('Salary').agg({"Salary":['mean','std','median']})
display(salary_group)
```

Standard Scaler based on its mean and standard deviation
z=x-u/sigma

```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
std_scaler = StandardScaler()


titanic_df = sns.load_dataset('titanic').dropna()

display(titanic_df)

display(titanic_df)
titanic_df['age']=std_scaler.fit_transform(np.array(titanic_df['age']).reshape(-1,1))
```

```python
titanic_df['age'].hist(bins=5)
plt.title("Distribution of Age After Standard Scaling")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

MinMaxScaler Xnew= ( X - Xmin )/ ( Xmax - Xmin )

<b>
This formula essentially resizes your data to fit within the range of 0 to 1.
</b>

```python
# titanic_df['Age']=
from sklearn.preprocessing import MinMaxScaler

titanic_df = sns.load_dataset('titanic').dropna()

min_max=MinMaxScaler()

display(titanic_df)

titanic_df['fare']=min_max.fit_transform(np.array(titanic_df['fare']).reshape(-1,1))

display(titanic_df)


titanic_df['fare'].hist()

plt.title("Fare Distribution based on Min Max Scaler")
plt.xlabel("Fare (min max unit)")
plt.ylabel("Frequency")
```

Robust Scaling when we have outliers with in our data.

<b>
 it scales data according to its IQR (Inter Quartile Range). Effectively, it's robust against outliers since it uses the IQR, and outliers fall outside the IQR.
</b>

```python
from sklearn.preprocessing import RobustScaler

titanic_df = sns.load_dataset('titanic').dropna()

# min_max=MinMaxScaler()
robust=RobustScaler()

display(titanic_df)

titanic_df['fare']=robust.fit_transform(np.array(titanic_df['fare']).reshape(-1,1))

display(titanic_df)
```

```python
titanic_df['fare'].hist()

plt.title("Fare Distribution based on Robust Scaling")
plt.xlabel("Fare (robust unit)")
plt.ylabel("Frequency")
plt.show()
```
