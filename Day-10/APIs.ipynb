Implement a program that makes multiple API requests and saves the data in a CSV file

```python
# 03f7baf9bed24035bc5211634242305
import requests
import csv
city=['London','Pennsylvania','New York City','Ahmedabad','Gandhinagar']


weather_data=[]

for i in city:
    response=requests.get(f"http://api.weatherapi.com/v1/current.json?key=03f7baf9bed24035bc5211634242305&q={i}&aqi=no")
    if response:
        response=response.json()
        weather_info={
            'city':response["location"]['region'],
            'country':response["location"]['country'],
            'temperature_calcius':response['current']['temp_c'],
            'weather':response['current']['condition']['text']
        }
        weather_data.append(weather_info)
    # print(response.json())
csv_file='weather_data.csv'
with open(csv_file,'w',newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['city', 'country', 'temperature_calcius', 'weather'])
    writer.writeheader()
    for row in weather_data:
        writer.writerow(row)


print(f"Data saved in {csv_file}")
```
