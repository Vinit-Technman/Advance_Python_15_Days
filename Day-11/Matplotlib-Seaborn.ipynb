Create a scatter plot with multiple groups using Matplotlib

```python
import numpy as np

# Group 1
x1 = np.random.rand(20)
y1 = np.random.rand(20)

# Group 2
x2 = np.random.rand(10)
y2 = np.random.rand(10)

# Group 3
x3 = np.random.rand(50)
y3 = np.random.rand(50)

plt.figure(figsize=(8,4))
plt.scatter(x1,y1,c='red')
plt.scatter(x2,y2,c='yellow')
plt.scatter(x3,y3,c='blue')

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

performance of two teams over time

Write a Python program that uses Matplotlib to create a 
histogram to show the distribution of exam score

```python
import matplotlib.pyplot as plt
import numpy as np

exam_scores=np.random.randint(50,100,size=400)


plt.hist(exam_scores,bins=15,edgecolor='black')
plt.title("Distribution of Exam Score")
plt.xlabel("Exam Score")
plt.ylabel("Freq")

# mean
mean_score = sum(exam_scores) / len(exam_scores)
plt.axvline(mean_score, color='r', linestyle='dashed', linewidth=1)
plt.show()
```

Make a subplot with a line plot and a bar plot to visualize 
both the trend and the comparison of value

```python
x1=np.random.randint(20,500,12)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
x2=np.random.randint(100,500,12)

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))

# line
axes[0].plot(months,x1,label='Values')
axes[0].set_xlabel('Months')
axes[0].set_ylabel('Values')
axes[0].legend()
axes[0].grid(True)

# bar
axes[1].bar(months,x1,label='Values')
axes[1].set_xlabel('Months')
axes[1].set_ylabel('Values')
axes[1].legend()
axes[1].grid(True)
plt.tight_layout()
plt.show()
```
